```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/Awesome-Machine-Learning/Machine-Learning-Classifications/master/03%20Classification%20Algorithms/data/df.csv')
df.head()
```

```python
import matplotlib.pyplot as plt

def plot(df):
  for cols in df.columns:
    df[cols].value_counts().plot(kind='bar')
    plt.show()
```

```python
plot(df)
```

```python
cat = ['age_bin','capital_gl_bin','education_bin','education_num_bin','hours_per_week_bin','msr_bin','occupation_bin','race_sex_bin']

df = pd.get_dummies(df,sparse=False,columns=cat,drop_first=True)
```

```python
dummies = df[['age_bin_b. 26-30 & 71-100',
       'age_bin_c. 31-35 & 61-70', 'age_bin_d. 36-40 & 56-60',
       'age_bin_e. 40-55', 'capital_gl_bin_b. < 0', 'capital_gl_bin_c. > 0',
       'education_bin_b. Mid', 'education_bin_c. Bachelors',
       'education_bin_d. Masters', 'education_bin_e. High',
       'education_num_bin_b. 9-12', 'education_num_bin_c. 13',
       'education_num_bin_d. 14', 'education_num_bin_e. 15+',
       'hours_per_week_bin_b. 31-40', 'hours_per_week_bin_c. 71-100',
       'hours_per_week_bin_d. 41-50 & 61-70', 'hours_per_week_bin_e. 51-60',
       'msr_bin_b. Mid', 'msr_bin_c. High', 'occupation_bin_b. Mid - Low',
       'occupation_bin_c. Mid - Mid', 'occupation_bin_d. Mid - High',
       'occupation_bin_e. High', 'race_sex_bin_b. Mid',
       'race_sex_bin_c. High']]
```

```python
dummies.info()
```

```python
dummies = dummies.astype(int)
```

```python
df = df.drop(columns=dummies)
```

```python
df = pd.concat([df,dummies],axis=1)
```

```python
df.head()
```

```python
train = df[df['flag']=='train']
test = df[df['flag']=='test']
```

```python
train = train.drop(columns='flag')
test = test.drop(columns='flag')
```

```python
train.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder

oe = OrdinalEncoder()
train[['workclass_bin']] = oe.fit_transform(train[['workclass_bin']])
test[['workclass_bin']] = oe.fit_transform(test[['workclass_bin']])
```

```python
train
```

```python
x_train = train.drop(columns='y')
y_train = train['y']

x_test = test.drop(columns='y')
y_test = test['y']
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(max_iter=100000)
lr.fit(x_train,y_train)
```

```python
pred = lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score

score = accuracy_score(y_test,pred)*100
score
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
knn.fit(x_train,y_train)
```

```python
knn_pred = knn.predict(x_test)
```

```python
score = accuracy_score(y_test,knn_pred)*100
score
```

```python
from sklearn.svm import SVC

svc = SVC()
svc.fit(x_train,y_train)
```

```python
svc_pred = svc.predict(x_test)
```

```python
score = accuracy_score(y_test,svc_pred)*100
score
```

```python
from sklearn.naive_bayes import GaussianNB

nb =  GaussianNB()
nb.fit(x_train, y_train)
nb_pred=nb.predict(x_test)
```

```python
score = accuracy_score(y_test,nb_pred)*100
score
```

```python
from sklearn.tree import DecisionTreeClassifier

dtc = DecisionTreeClassifier()
dtc.fit(x_train,y_train)
dtc_pred = dtc.predict(x_test)
```

```python
score = accuracy_score(y_test,dtc_pred)*100
score
```

```python
from sklearn.linear_model import SGDClassifier

sgd = SGDClassifier()
sgd.fit(x_train,y_train)
sgd_pred = sgd.predict(x_test)
```

```python
score = accuracy_score(y_test,sgd_pred)*100
score
```
